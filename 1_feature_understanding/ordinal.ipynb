# 尺度分析

##### Metric
- Nominal:
按照「名稱」做分類


- Ordinal:
可以「計數」，可以做「排序」與「比較」。但是數值之間的計算是沒有意義的。

##### Non-metric
- Interval:
數值可以做排序跟比較外，還可以做「加減」。
可以計算「算術平均數」及「標準差」。

- Ratio:
有「絕對零點」，除了加減運算外，還可以「乘除」。

## Ordinal
資料來源: https://data.sfgov.org/Transportation/2013-SFO-Customer-Survey-Data-Set-Dictionary/mjr8-p6m5

```python
#匯入新資料
customer = pd.read_csv('./data/2013_SFO_Customer_Survey.csv')
customer.shape
```

```python
#Q7A_ART是關於藝術品和展覽。(1:不可接受、2:低於平均、3:平均、4:不錯、5:特別好、6:從未有人使用或參觀過、0:空)

art_ratings = customer['Q7A_ART']
art_ratings.describe()
```

```python
#只考慮1~5
art_ratings = art_ratings[(art_ratings>=1) & (art_ratings<=5)]
```

```python
#將數值轉換成字串
art_ratings = art_ratings.astype(str)
art_ratings.describe()
```

```python
#視覺化，圓餅圖
art_ratings.value_counts().plot(kind='pie')
```

```python
#長條圖
art_ratings.value_counts().plot(kind='bar')
```

```python
#盒鬚圖
art_ratings.value_counts().plot(kind='box')
```
